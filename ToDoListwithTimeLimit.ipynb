```python
import os

#Function to save task in a filename
def save_task(todo_tasks, filename):
  with open(filename, 'w') as file:
    for task in todo_tasks:
      file.write(f"{task['task_name']},{task['task_timelimit']} \n")
#function to load task from file
def load_task(filename):
  tasks = []
  if os.path.exists(filename):
    with open(filename, 'r') as file:
      for line in file:
        task_name, task_timelimit = line.strip().split(',')
        tasks.append({"task_name": task_name, "task_timelimit": task_timelimit})
  return tasks

#function to create a task to do list with time limit
def create_newTask(todo_tasks, task_name, task_timelimit):
  task = {"task_name": task_name, "task_timelimit": task_timelimit}
  # todo_tasks[task_name] = task_timelimit
  todo_tasks.append(task)
  print("Task added successfully.")

#function to delete a task from the LIST
def del_oldtask(todo_tasks, task_index):
  if task_index >= 0 and task_index < len(todo_tasks):
    deleted_task = todo_tasks.pop(task_index)
    print(f"todo_tasks '{deleted_task['task_name']}' deleted.")
  else:
    print("Invalid task index.")

# Function to update a task
def update_oldtask(todo_tasks, task_index, new_task_name, new_timelimit):
    if task_index >= 0 and task_index < len(todo_tasks):
      todo_tasks[task_index]["task_name"] = new_task_name
      todo_tasks[task_index]["task_timelimit"] = new_timelimit
      print("Task updated successfully.")
    else:
        print("Invalid task index.")

# Function to view all tasks
def view_tasks(tasks):
    print("To-Do List:")
    for index, task in enumerate(tasks, start=1):
        print(f"{index}. Task: {task['task_name']} | Time Limit: {task['task_timelimit']}")

# def view_tasks(todo_tasks):
#     print("TO DO LIST: \n")
#     for task_name, task_timelimit in todo_tasks.items():
#         print(f"{task_name}: {task_timelimit}")

def main():
  filename = "tasklist.txt"
  todo_tasks_list = load_task(filename)

  while True:
    print("\nTo-Do Task List Management Database")
    print("Select: 1. To Add  New Task")
    print("Select: 2. To View All Tasks")
    print("Select: 3. To Update Old Task")
    print("Select: 4. To Delete Old Task")
    print("Select: 5. To Exit Application")

    select = input("Enter your Select: \n")
    if select == '1':
      task_name = input("Enter task name: ")
      timelimit = input("Enter time limit: ")
      create_newTask(todo_tasks_list, task_name, timelimit)
      save_task(todo_tasks_list, filename)
    elif select == '2':
      view_tasks(todo_tasks_list)
    elif select == '3':
      task_index = int(input("Enter the index of the task to update: ")) - 1
      new_task_name = input("Enter new task name: ")
      new_timelimit = input("Enter new time limit: ")
      update_oldtask(todo_tasks_list, task_index, new_task_name, new_timelimit)
      save_task(todo_tasks_list, filename)
    elif select == '4':
      task_index = int(input("Enter the index of the task to delete: ")) - 1
      del_oldtask(todo_tasks_list, task_index)
      save_task(todo_tasks_list, filename)
    elif select == '5':
      print("Goodbye!")
      break
    else:
      print("Invalid choice. Please try again.")



if __name__ == "__main__":
    main()


```
